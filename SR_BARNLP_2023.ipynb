# SISTEMAS DE RECOMENDAÇÃO

* Sistemas de Recomendação surgiram pela primeira vez na década de 90 com intuito de minimizar os problemas com a grande quantidade de informações (MAES; SHARADANAND, 1995). Este tipo de sistema tem sido utilizado largamente pro empresas como as de em E-Commerce (Amazon, Mercado Livre, Google), Provedores de streaming (Netflix, Amazon Prime Video, Globo Play) entre outros.

### OBJETIVO DO SR

*   Um sistema que prediz avaliações ou preferências.
*   Com base em seu comportamento anterior ou de outras pessoas.
*   Recomendando lista de itens

SR2.png

* Os sistemas de recomendação utilizam fontes distintas de informação para fornecer aos usuários recomendações de itens, visando equilibrar elementos como novidade, precisão, dispersão e estabilidade nas recomendações. Desta forma, são empregadas diversas técnicas de recomendação, que utilizam diferentes tipos de abordagem.
* Os tipos de sistemas de recomendação que podem ser implementados segundo BURKE, (2007), são divididos em quatro classes distintas:
> * baseadas em conteúdo
> * filtragem colaborativa
> * demográfico
> * baseada em conhecimento.

image.png


* Ainda outro estudo anterior, BURKE, (2002) apresentou também outras duas técnicas baseada na utilidade e o modelo híbrido, tendo como objetivo a junção de duas técnicas ou mais, necessitando que se utilize uma grande quantidade de dados para que a obtenção de recomendações de qualidade possa ser feita.
* O uso de machine learning é uma das abordagens tecnologicas possiveis para implementarum sistema de recomendação. O machine learning, ou aprendizado de máquinas, é a área da inteligência artificial relacionada à busca de um conjunto de regras e procedimentos para permitir que as máquinas possam agir e tomar decisões baseadas em dados, ao invés de serem explicitamente programadas para realizar uma determinada tarefa.


# MACHINE LEARNING
* Machine Learning é uma área da inteligência artificial que vem sendo estudada e utilizada para simular o aprendizado humano, permitir e obter conhecimento do mundo real a nível computacional.
* Para entender o que é machine learning, é preciso conhecer suas as principais técnicas e modelos, porém, ele é comumente dividido em três principais categorias: aprendizado supervisionado, aprendizado não-supervisionado e aprendizado por reforço.

>* Supervisionado: a categoria mais utilizada de machine learning usa um algoritmo que precisa de exemplos rotulados para desempenhar suas tarefas. Ou seja, é preciso programar a pergunta que a máquina deve responder, cada quesito a ser avaliado e as opções possíveis de resposta.

>* Não-supervisionado: esses algoritmos visam obter uma melhor representação dos dados, como em uma segmentação/classificação de clientes, e não usam rótulos pré-definidos.

>* Aprendizado por reforço: máquinas dotadas desse tipo de aprendizado conseguem desenvolver uma política de ações visando uma determinada recompensa. Aqui entram, por exemplo, máquinas que aprendem a jogar xadrez e tomam decisões embasadas em cada estado do jogo para maximizar a recompensa final (vencer a partida).

## Filtro Baseado em Conteúdo
A técnica de filtro baseado em conteúdo (NIEMANN; WOLPERS, 2015) (BEEL et al., 2016) (BOURKOUKOU; ELBACHARI; ELADNANI, 2017) (SATTAR; GHAZANFAR; IQBAL, 2017), busca-se identificar itens que possam ser de interesse do usuário tomando como base para a decisão outros itens que o usuário já tenha consumido e avaliado (PAZZANI; BILLSUS, 2007), (AGGARWAL, 2016). Parte do princípio de que um usuário tem a tendência de se interessar por elementos similares, que tenham alguma relação com itens consumidos no passado (LOPS; DE GEMMIS; SEMERARO, 2011).

## Recomendação 2
* Selecionar os filmes por ordem de nota (media) descrescente
* Filtrar os com melhor nota


```python
notas.head()
```

```python
medias = notas.groupby("filmeId").mean()["nota"]
medias.head(10)
```

```python
filmes["media"]=medias
filmes.head(10)
```

```python
filmes.sort_values("media",ascending=False).head(10)
```

```python
filmes_com_relevancia = filmes.query("quantidade_rating >= 50").sort_values("media",ascending=False)
filmes_com_relevancia.head(10)
```
